## Варіант 4

### Постановка задачі

**Функція:**

$$y = \frac{e^x - e^{-x}}{2} + \ln(x^2 + 1)$$

**ОДЗ:** Функція визначена для всіх $x$ (гіперболічний синус + логарифм від завжди додатного виразу).

**Додаткова умова:** Знайти добуток елементів списку з парними індексами (0, 2, 4, ...).

### Лістінг програми

```python
import math

# Константа точності
EPS = 1e-9

# ============================================================
# ПРАКТИЧНА РОБОТА №2 - ВАРІАНТ 4
# ============================================================
print("=" * 60)
print("ПРАКТИЧНА РОБОТА №2: ІТЕРАЦІЙНІ ОБЧИСЛЕННЯ")
print("Варіант 4")
print("=" * 60)

print("\nФункція: y = (e^x - e^(-x))/2 + ln(x² + 1)")
print("ОДЗ: функція визначена для всіх x")
print("-" * 60)

# ============================================================
# ЧАСТИНА 1: ВВЕДЕННЯ ВИХІДНИХ ДАНИХ
# ============================================================
print("\n--- ВВЕДЕННЯ ВИХІДНИХ ДАНИХ ---")
x_start = float(input("Введіть початкове значення x_start: "))
x_end = float(input("Введіть кінцеве значення x_end: "))
h = float(input("Введіть крок табуляції h: "))

# ============================================================
# ЧАСТИНА 2: ТАБУЛЯЦІЯ ФУНКЦІЇ
# ============================================================
print("\n--- ТАБУЛЯЦІЯ ФУНКЦІЇ ---")

# Ініціалізація списку
y_list = []

# Виведення заголовка таблиці
print(f"\n{'№':>5} | {'x':>12} | {'y':>15}")
print("-" * 40)

# Лічильник точок
point_number = 0

# Цикл табуляції
x = x_start
while x <= x_end + EPS:
    # Обчислення функції: y = sinh(x) + ln(x² + 1)
    # sinh(x) = (e^x - e^(-x)) / 2
    y = (math.exp(x) - math.exp(-x)) / 2 + math.log(x * x + 1)
    
    # Збереження значення у список
    y_list.append(y)
    
    # Виведення результату
    point_number += 1
    print(f"{point_number:5d} | {x:12.6f} | {y:15.8f}")
    
    # Перехід до наступної точки
    x += h

print("-" * 40)
print(f"Загальна кількість точок: {len(y_list)}")

# ============================================================
# ЧАСТИНА 3: ОБРОБКА СПИСКУ
# ============================================================
print("\n--- ОБРОБКА СПИСКУ ЗНАЧЕНЬ ---")

if len(y_list) > 0:
    # Пошук максимуму та мінімуму
    max_y = y_list[0]
    min_y = y_list[0]
    
    for y in y_list:
        if y > max_y:
            max_y = y
        if y < min_y:
            min_y = y
    
    # Обчислення суми та середнього
    total = 0.0
    for y in y_list:
        total += y
    
    average = total / len(y_list)
    
    # Виведення основних результатів
    print(f"\n{'=' * 60}")
    print("ОСНОВНА СТАТИСТИКА:")
    print(f"{'=' * 60}")
    print(f"Максимальне значення y: {max_y:.8f}")
    print(f"Мінімальне значення y: {min_y:.8f}")
    print(f"Сума всіх елементів: {total:.8f}")
    print(f"Середнє арифметичне: {average:.8f}")
    
    # ============================================================
    # ДОДАТКОВА УМОВА: добуток елементів з парними індексами
    # ============================================================
    print(f"\n{'=' * 60}")
    print("ДОДАТКОВА УМОВА:")
    print("Добуток елементів списку з парними індексами (0, 2, 4, ...)")
    print(f"{'=' * 60}")
    
    product_even = 1.0
    count_even = 0
    
    print(f"\n{'Індекс':>8} | {'Значення y':>15} | {'Статус':>15}")
    print("-" * 45)
    
    for i in range(len(y_list)):
        y = y_list[i]
        if i % 2 == 0:
            product_even *= y
            count_even += 1
            status = "Парний індекс"
        else:
            status = "-"
        
        print(f"{i:8d} | {y:15.8f} | {status:>15}")
    
    print("-" * 45)
    
    print(f"\nКількість елементів з парними індексами: {count_even}")
    print(f"Добуток елементів з парними індексами: {product_even:.8f}")

else:
    print("Список порожній. Немає даних для обробки.")

print(f"\n{'=' * 60}")
print("ПРОГРАМУ ЗАВЕРШЕНО")
print(f"{'=' * 60}")
```

## Варіант 6

### Постановка задачі

**Функція:**

$$y = \tan(x) - \frac{1}{x^2 + 1}$$

**ОДЗ:** $\cos(x) \neq 0$, тобто $x \neq \frac{\pi}{2} + \pi k$, де $k$ — ціле число.

**Додаткова умова:** Знайти кількість елементів списку, абсолютне значення яких менше 1.

### Лістінг програми

```python
import math

# Константа точності
EPS = 1e-9

# ============================================================
# ПРАКТИЧНА РОБОТА №2 - ВАРІАНТ 6
# ============================================================
print("=" * 60)
print("ПРАКТИЧНА РОБОТА №2: ІТЕРАЦІЙНІ ОБЧИСЛЕННЯ")
print("Варіант 6")
print("=" * 60)

print("\nФункція: y = tan(x) - 1/(x² + 1)")
print("ОДЗ: cos(x) ≠ 0, тобто x ≠ π/2 + πk")
print("-" * 60)

# ============================================================
# ЧАСТИНА 1: ВВЕДЕННЯ ВИХІДНИХ ДАНИХ
# ============================================================
print("\n--- ВВЕДЕННЯ ВИХІДНИХ ДАНИХ ---")
x_start = float(input("Введіть початкове значення x_start: "))
x_end = float(input("Введіть кінцеве значення x_end: "))
h = float(input("Введіть крок табуляції h: "))

# ============================================================
# ЧАСТИНА 2: ТАБУЛЯЦІЯ ФУНКЦІЇ
# ============================================================
print("\n--- ТАБУЛЯЦІЯ ФУНКЦІЇ ---")

# Ініціалізація списку
y_list = []

# Виведення заголовка таблиці
print(f"\n{'№':>5} | {'x':>12} | {'y':>15} | {'Статус':>15}")
print("-" * 55)

# Лічильник точок
point_number = 0

# Цикл табуляції
x = x_start
while x <= x_end + EPS:
    # Перевірка ОДЗ: cos(x) ≠ 0
    if abs(math.cos(x)) > EPS:
        # Обчислення функції
        y = math.tan(x) - 1 / (x * x + 1)
        
        # Збереження значення у список
        y_list.append(y)
        
        # Виведення результату
        point_number += 1
        print(f"{point_number:5d} | {x:12.6f} | {y:15.8f} | {'OK':>15}")
    else:
        print(f"{'---':>5} | {x:12.6f} | {'---':>15} | {'cos(x) = 0':>15}")
    
    # Перехід до наступної точки
    x += h

print("-" * 55)
print(f"Загальна кількість обчислених точок: {len(y_list)}")

# ============================================================
# ЧАСТИНА 3: ОБРОБКА СПИСКУ
# ============================================================
print("\n--- ОБРОБКА СПИСКУ ЗНАЧЕНЬ ---")

if len(y_list) > 0:
    # Пошук максимуму та мінімуму
    max_y = y_list[0]
    min_y = y_list[0]
    
    for y in y_list:
        if y > max_y:
            max_y = y
        if y < min_y:
            min_y = y
    
    # Обчислення суми та середнього
    total = 0.0
    for y in y_list:
        total += y
    
    average = total / len(y_list)
    
    # Виведення основних результатів
    print(f"\n{'=' * 60}")
    print("ОСНОВНА СТАТИСТИКА:")
    print(f"{'=' * 60}")
    print(f"Максимальне значення y: {max_y:.8f}")
    print(f"Мінімальне значення y: {min_y:.8f}")
    print(f"Сума всіх елементів: {total:.8f}")
    print(f"Середнє арифметичне: {average:.8f}")
    
    # ============================================================
    # ДОДАТКОВА УМОВА: кількість елементів з |y| < 1
    # ============================================================
    print(f"\n{'=' * 60}")
    print("ДОДАТКОВА УМОВА:")
    print("Кількість елементів списку, абсолютне значення яких менше 1")
    print(f"{'=' * 60}")
    
    count_abs_lt_1 = 0
    
    print(f"\n{'№':>5} | {'Значення y':>15} | {'|y|':>12} | {'|y| < 1?':>10}")
    print("-" * 50)
    
    for i in range(len(y_list)):
        y = y_list[i]
        abs_y = abs(y)
        if abs_y < 1 - EPS:
            count_abs_lt_1 += 1
            status = "Так"
        else:
            status = "Ні"
        
        print(f"{i+1:5d} | {y:15.8f} | {abs_y:12.8f} | {status:>10}")
    
    print("-" * 50)
    
    print(f"\nКількість елементів з |y| < 1: {count_abs_lt_1}")

else:
    print("Список порожній. Немає даних для обробки.")

print(f"\n{'=' * 60}")
print("ПРОГРАМУ ЗАВЕРШЕНО")
print(f"{'=' * 60}")
```

## Варіант 8

### Постановка задачі

**Функція:**

$$y = \frac{\ln(x^2 + 1)}{\cos(x) + 2}$$

**ОДЗ:** $\cos(x) + 2 \neq 0$ — виконується завжди, оскільки $\cos(x) \geq -1$, тому $\cos(x) + 2 \geq 1 > 0$.

**Додаткова умова:** Знайти добуток елементів списку, значення яких знаходяться в інтервалі $[-2, 2]$.

### Лістінг програми

```python
import math

# Константа точності
EPS = 1e-9

# ============================================================
# ПРАКТИЧНА РОБОТА №2 - ВАРІАНТ 8
# ============================================================
print("=" * 60)
print("ПРАКТИЧНА РОБОТА №2: ІТЕРАЦІЙНІ ОБЧИСЛЕННЯ")
print("Варіант 8")
print("=" * 60)

print("\nФункція: y = ln(x² + 1) / (cos(x) + 2)")
print("ОДЗ: функція визначена для всіх x (знаменник завжди > 0)")
print("-" * 60)

# ============================================================
# ЧАСТИНА 1: ВВЕДЕННЯ ВИХІДНИХ ДАНИХ
# ============================================================
print("\n--- ВВЕДЕННЯ ВИХІДНИХ ДАНИХ ---")
x_start = float(input("Введіть початкове значення x_start: "))
x_end = float(input("Введіть кінцеве значення x_end: "))
h = float(input("Введіть крок табуляції h: "))

# ============================================================
# ЧАСТИНА 2: ТАБУЛЯЦІЯ ФУНКЦІЇ
# ============================================================
print("\n--- ТАБУЛЯЦІЯ ФУНКЦІЇ ---")

# Ініціалізація списку
y_list = []

# Виведення заголовка таблиці
print(f"\n{'№':>5} | {'x':>12} | {'y':>15}")
print("-" * 40)

# Лічильник точок
point_number = 0

# Цикл табуляції
x = x_start
while x <= x_end + EPS:
    # Обчислення функції (ОДЗ завжди виконана)
    y = math.log(x * x + 1) / (math.cos(x) + 2)
    
    # Збереження значення у список
    y_list.append(y)
    
    # Виведення результату
    point_number += 1
    print(f"{point_number:5d} | {x:12.6f} | {y:15.8f}")
    
    # Перехід до наступної точки
    x += h

print("-" * 40)
print(f"Загальна кількість точок: {len(y_list)}")

# ============================================================
# ЧАСТИНА 3: ОБРОБКА СПИСКУ
# ============================================================
print("\n--- ОБРОБКА СПИСКУ ЗНАЧЕНЬ ---")

if len(y_list) > 0:
    # Пошук максимуму та мінімуму
    max_y = y_list[0]
    min_y = y_list[0]
    
    for y in y_list:
        if y > max_y:
            max_y = y
        if y < min_y:
            min_y = y
    
    # Обчислення суми та середнього
    total = 0.0
    for y in y_list:
        total += y
    
    average = total / len(y_list)
    
    # Виведення основних результатів
    print(f"\n{'=' * 60}")
    print("ОСНОВНА СТАТИСТИКА:")
    print(f"{'=' * 60}")
    print(f"Максимальне значення y: {max_y:.8f}")
    print(f"Мінімальне значення y: {min_y:.8f}")
    print(f"Сума всіх елементів: {total:.8f}")
    print(f"Середнє арифметичне: {average:.8f}")
    
    # ============================================================
    # ДОДАТКОВА УМОВА: добуток елементів в інтервалі [-2, 2]
    # ============================================================
    print(f"\n{'=' * 60}")
    print("ДОДАТКОВА УМОВА:")
    print("Добуток елементів списку, значення яких в інтервалі [-2, 2]")
    print(f"{'=' * 60}")
    
    product_interval = 1.0
    count_interval = 0
    
    print(f"\n{'№':>5} | {'Значення y':>15} | {'В інтервалі?':>15}")
    print("-" * 45)
    
    for i in range(len(y_list)):
        y = y_list[i]
        if -2 - EPS <= y <= 2 + EPS:
            product_interval *= y
            count_interval += 1
            status = "Так"
        else:
            status = "Ні"
        
        print(f"{i+1:5d} | {y:15.8f} | {status:>15}")
    
    print("-" * 45)
    
    print(f"\nКількість елементів в інтервалі [-2, 2]: {count_interval}")
    if count_interval > 0:
        print(f"Добуток елементів в інтервалі [-2, 2]: {product_interval:.8f}")
    else:
        print("Немає елементів в заданому інтервалі.")

else:
    print("Список порожній. Немає даних для обробки.")

print(f"\n{'=' * 60}")
print("ПРОГРАМУ ЗАВЕРШЕНО")
print(f"{'=' * 60}")
```
