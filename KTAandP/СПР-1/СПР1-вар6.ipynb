## Постановка задачі

Написати програму, яка запитує значення x і обчислює кусково-задану функцію варіанта 6:

$$
y = \begin{cases}
5^{-\sin x} \cdot \sqrt{x+2}, & x \leq -2 \\
\frac{\sin^5 x - \pi x}{\operatorname{ctg} x}, & -2 < x < 0 \\
\sqrt[6]{x} + 0.628x, & 0 < x \leq 4 \\
\operatorname{ctg} x - e^x, & x > 4.2
\end{cases}
$$

**Примітка:** Функція не визначена на інтервалах $x = 0$ та $(4; 4.2]$.

## Визначення ОДЗ функції

Для кожного інтервалу визначимо обмеження:

1. **Інтервал** $x \leq -2$: $y = 5^{-\sin x} \cdot \sqrt{x+2}$
   - $x + 2 \geq 0$, тобто $x \geq -2$
   - На межі інтервалу при $x = -2$: $\sqrt{-2+2} = \sqrt{0} = 0$

2. **Інтервал** $-2 < x < 0$: $y = \frac{\sin^5 x - \pi x}{\operatorname{ctg} x}$
   - $\sin x \neq 0$ (для визначеності котангенса)
   - $\operatorname{ctg} x = \frac{\cos x}{\sin x}$

3. **Інтервал** $0 < x \leq 4$: $y = \sqrt[6]{x} + 0.628x$
   - $x > 0$ для визначеності кореня шостого степеня
   - Обмежень немає, оскільки $x > 0$ в цьому інтервалі

4. **Інтервал** $x > 4.2$: $y = \operatorname{ctg} x - e^x$
   - $\sin x \neq 0$ (для визначеності котангенса)

## Лістинг програми

````python
import math

t = 1e-9

x = float(input("Input x = "))

# Перевірка чи x = 0 (невизначена точка)
if abs(x) < t:
    print("Функція не визначена в точці x = 0")
# Перевірка чи x у невизначеному інтервалі (4; 4.2]
elif (4 - t) < x <= (4.2 + t):
    print("Функція не визначена на інтервалі (4; 4.2]")
else:
    # Інтервал 1: x <= -2
    if x <= (-2 + t):
        radicand = x + 2
        
        # Перевірка radicand >= 0
        if radicand > -t:
            # Обробка випадку radicand = 0
            if radicand < t:
                sqrt_part = 0
            else:
                sqrt_part = math.sqrt(radicand)
            
            power_part = 5 ** (-math.sin(x))
            y = power_part * sqrt_part
            print(f"x = {x}, y1 = {y:.6f}")
        else:
            print("Порушено ОДЗ: вираз під коренем від'ємний")
    
    # Інтервал 2: -2 < x < 0
    elif (-2 + t) < x < -t:
        # Перевірка sin(x) != 0 для котангенса
        if abs(math.sin(x)) > t:
            cotangent = math.cos(x) / math.sin(x)
            numerator = math.sin(x) ** 5 - math.pi * x
            y = numerator / cotangent
            print(f"x = {x}, y2 = {y:.6f}")
        else:
            print("Порушено ОДЗ: sin(x) = 0, котангенс не визначений")
    
    # Інтервал 3: 0 < x <= 4
    elif t < x <= (4 - t):
        # Корінь 6-го степеня від x + додатково 0.628x
        y = x ** (1.0 / 6.0) + 0.628 * x
        print(f"x = {x}, y3 = {y:.6f}")
    
    # Інтервал 4: x > 4.2
    elif x > (4.2 + t):
        # Перевірка sin(x) != 0 для котангенса
        if abs(math.sin(x)) > t:
            cotangent = math.cos(x) / math.sin(x)
            y = cotangent - math.exp(x)
            print(f"x = {x}, y4 = {y:.6f}")
        else:
            print("Порушено ОДЗ: sin(x) = 0, котангенс не визначений")
````

## Результати роботи програми

Тестові приклади для різних інтервалів:

### Тест 1: Інтервал 1 (x ≤ -2), межа x = -2
```
Input x = -2.0
x = -2.0, y1 = 0.000000
```

**Перевірка:**
- $\sin(-2) \approx -0.9093$
- $5^{-(-0.9093)} = 5^{0.9093} \approx 4.5175$
- $\sqrt{-2+2} = \sqrt{0} = 0$
- $y = 4.5175 \cdot 0 = 0$ ✓

### Тест 2: Інтервал 1 (x ≤ -2), x < -2 - порушення ОДЗ
```
Input x = -2.5
Порушено ОДЗ: вираз під коренем від'ємний
```

**Перевірка:**
- Підкореневий вираз: $x + 2 = -2.5 + 2 = -0.5 < 0$
- Програма коректно виявляє порушення ОДЗ ✓

### Тест 3: Інтервал 2 (-2 < x < 0)
```
Input x = -1.0
x = -1.0, y2 = -4.235691
```

**Перевірка:**
- $\sin(-1) \approx -0.8415$
- $\cos(-1) \approx 0.5403$
- $\operatorname{ctg}(-1) = \frac{0.5403}{-0.8415} \approx -0.6421$
- $\sin^5(-1) \approx (-0.8415)^5 \approx -0.4212$
- Чисельник: $\sin^5(-1) - \pi \cdot (-1) = -0.4212 + 3.1416 = 2.7204$
- $y = \frac{2.7204}{-0.6421} \approx -4.235691$ ✓

### Тест 4: x = 0 (невизначена точка)
```
Input x = 0.0
Функція не визначена в точці x = 0
```

### Тест 5: Інтервал 3 (0 < x ≤ 4)
```
Input x = 2.0
x = 2.0, y3 = 2.378462
```

**Перевірка:**
- $y = \sqrt[6]{2} + 0.628 \cdot 2 = 2^{1/6} + 1.256 \approx 1.1225 + 1.256 \approx 2.378462$ ✓

### Тест 6: Невизначений інтервал (4 < x ≤ 4.2)
```
Input x = 4.1
Функція не визначена на інтервалі (4; 4.2]
```

### Тест 7: Інтервал 4 (x > 4.2)
```
Input x = 5.0
x = 5.0, y4 = -148.708972
```

**Перевірка:**
- $\sin(5) \approx -0.9589$
- $\cos(5) \approx 0.2837$
- $\operatorname{ctg}(5) = \frac{0.2837}{-0.9589} \approx -0.2958$
- $e^5 \approx 148.4132$
- $y = -0.2958 - 148.4132 \approx -148.708972$ ✓

### Тест 8: Порушення ОДЗ (ctg(x) при x = 2π)
```
Input x = 6.283185307179586
Порушено ОДЗ: sin(x) = 0, котангенс не визначений
```

**Примітка:** Значення $x = 2\pi \approx 6.2832 > 4.2$, що належить інтервалу 4, де формула містить котангенс. Оскільки $\sin(2\pi) \approx 0$ (через машинну похибку дорівнює приблизно $-2.4493 \times 10^{-16}$, що менше за $t = 10^{-9}$), програма коректно виявляє порушення ОДЗ.
