## Постановка задачі

Написати програму, яка запитує значення x і обчислює кусково-задану функцію варіанта 4:

$$
y = \begin{cases}
x^3 + e^{-1.3x}, & x \leq 0 \\
e^x \sin x + 7.3, & 0 < x < 0.5 \\
\frac{\tan x}{\sqrt{5+x^2}}, & 0.5 \leq x < 1.4 \\
\frac{x^7}{11\pi - 0.5x^2}, & x > 1.7
\end{cases}
$$

**Примітка:** Функція не визначена на інтервалі $[1.4; 1.7]$.

## Визначення ОДЗ функції

Для кожного інтервалу визначимо обмеження:

1. **Інтервал** $x \leq 0$: $y = x^3 + e^{-1.3x}$
   - Немає обмежень

2. **Інтервал** $0 < x < 0.5$: $y = e^x \sin x + 7.3$
   - Немає обмежень

3. **Інтервал** $0.5 \leq x < 1.4$: $y = \frac{\tan x}{\sqrt{5+x^2}}$
   - $5 + x^2 > 0$ (завжди виконується)
   - $\cos x \neq 0$ (для визначеності тангенса)

4. **Інтервал** $x > 1.7$: $y = \frac{x^7}{11\pi - 0.5x^2}$
   - $11\pi - 0.5x^2 \neq 0$
   - $x^2 \neq \frac{11\pi}{0.5} = 22\pi \approx 69.115$
   - $x \neq \pm\sqrt{22\pi} \approx \pm 8.313$

## Блок-схема алгоритму

```mermaid
flowchart TD
    A(["Початок"]) --> B["Введення x, t = 10⁻⁹"]
    B --> C{"1.4 - t < x < 1.7 + t"}

    C1 --> X(["Кінець"])
    C -- Ні --> D{"x ≤ t"}

    F -- Так --> F1{"abs(cos(x)) > t"}
    F1 -- Ні --> F2["Порушено ОДЗ: cos(x) = 0"]
    F2 --> X
    F1 -- Так --> F3["y = tan(x) / sqrt(5 + x²)"]

    G2 -- Так --> G4["y = x⁷ / denominator"]
    G4 --> OUT
```

## Лістинг програми

````python
import math

t = 1e-9

x = float(input("Input x = "))

# Перевірка чи x у невизначеному інтервалі [1.4; 1.7]
if (1.4 - t) < x < (1.7 + t):
    print("Функція не визначена на інтервалі [1.4; 1.7]")
else:
    # Інтервал 1: x <= 0
    if x <= t:
        y = x ** 3 + math.exp(-1.3 * x)
        print(f"x = {x}, y1 = {y:.6f}")
    
    # Інтервал 2: 0 < x < 0.5
    elif t < x < (0.5 - t):
        y = math.exp(x) * math.sin(x) + 7.3
        print(f"x = {x}, y2 = {y:.6f}")
    
    # Інтервал 3: 0.5 <= x < 1.4
    elif (0.5 - t) <= x < (1.4 - t):
        # Перевірка cos(x) != 0 для tan(x)
        if abs(math.cos(x)) > t:
            y = math.tan(x) / math.sqrt(5 + x * x)
            print(f"x = {x}, y3 = {y:.6f}")
        else:
            print("Порушено ОДЗ: cos(x) = 0, tan(x) не визначений")
    
    # Інтервал 4: x > 1.7
    elif x > (1.7 + t):
        denominator = 11 * math.pi - 0.5 * x * x
        
        # Перевірка ділення на нуль
        if abs(denominator) > t:
            y = (x ** 7) / denominator
            print(f"x = {x}, y4 = {y:.6f}")
        else:
            print("Порушено ОДЗ: ділення на нуль")
````
