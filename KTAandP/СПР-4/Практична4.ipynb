## 1.1 Об'єктно-орієнтоване програмування: базові поняття

**Об'єктно-орієнтоване програмування (ООП)** — парадигма програмування, в якій програма описується як взаємодія об'єктів.

**Об'єкт** — це сутність, що має:
- **Стан** (дані, які зберігаються в атрибутах)
- **Поведінку** (операції, які реалізуються методами)

**Клас** — це шаблон (опис, blueprint) для створення об'єктів. Клас визначає, які атрибути та методи матимуть об'єкти.

**Об'єкт (екземпляр)** — це конкретна реалізація класу з конкретними значеннями атрибутів.

### Основні принципи ООП:

1. **Інкапсуляція** — об'єднання даних та методів, що працюють з цими даними, в межах класу
2. **Наслідування** — можливість створення нових класів на основі існуючих
3. **Поліморфізм** — можливість об'єктів різних класів реагувати на однакові виклики по-різному
4. **Абстракція** — виділення суттєвих характеристик об'єкта та ігнорування несуттєвих

### Приклад концепції:



```python
# Клас — це шаблон
class Dog:
    def __init__(self, name, age):
        self.name = name  # атрибут
        self.age = age

    def __repr__(self):
        return f"Це собака на ім'я {self.name} віком {self.age} років"
    
    def bark(self):  # метод
        return f"{self.name} каже: Гав-гав!"

# Об'єкти — це конкретні екземпляри
dog1 = Dog("Рекс", 5)
dog2 = Dog("Бобік", 3)

print(dog1.bark())  # Рекс каже: Гав-гав!
print(dog2.bark())  # Бобік каже: Гав-гав!
```

## 1.2 Створення класів та об'єктів у Python

### Оголошення класу

Клас у Python оголошується ключовим словом `class`. За конвенцією, назви класів пишуться в PascalCase.



```python
class Example:
    """Документація класу (docstring)."""
    
    # Конструктор класу
    def __init__(self, a, b=0):
        """
        Ініціалізує об'єкт.
        
        Args:
            a: перший параметр
            b: другий параметр
        """
        self.a = a  # атрибут екземпляра
        self.b = b
    
    # Метод екземпляра
    def value(self):
        """Повертає суму a та b."""
        return self.a + self.b
    
    # Метод для виведення
    def __str__(self):
        return f"Example(a={self.a}, b={self.b})"

    def __repr__(self):
        return f"Instance of Example(a:{self.a}, b:{self.b})"
```

```python
ex1 = Example(1, 2)
```


### Створення об'єктів



```python
# Створення екземпляра класу
obj = Example(10, 20)

# Виклик методів
result = obj.value()  # 30
print(obj)  # Example(a=10, b=20)

# Доступ до атрибутів
print(obj.a)  # 10
```

## 1.3 Наслідування

**Наслідування** — механізм створення нового класу (похідного, дочірнього) на основі існуючого (базового, батьківського).

### Переваги наслідування:
- Повторне використання коду
- Розширення функціональності
- Створення ієрархії класів

### Синтаксис наслідування:



```python
class Base:
    """Базовий клас."""
    
    def __init__(self, x):
        self.x = x
    
    def calculate(self):
        return self.x * 2

    def power(self):
        return self.x ** self.x

class Child(Base):
    """Похідний клас."""
    
    def __init__(self, x, y):
        super().__init__(x)  # виклик конструктора базового класу
        self.y = y
    
    # Перевизначення методу (override)
    def calculate(self):
        return self.x * self.y
    
    # Новий метод
    def additional_method(self):
        return self.x + self.y

    def power(self):
        ans = super().power()
        print(f'{ans = }')
        res = ans // self.y
        print(f'{res = }')
        return res 
```

```python
base1 = Base(10)
```

```python
child1 = Child(20, 30)
```

```python
base1.calculate()
```

```python
child1.calculate()
```

```python
child1.additional_method()
```

```python
base1.additional_method()
```


### Використання `super()`

`super()` дозволяє викликати методи базового класу з похідного:



```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Звук тварини"

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # виклик __init__ базового класу
        self.breed = breed
    
    def speak(self):
        super().speak()  # виклик методу базового класу
        return f"{self.name} каже: Гав!"
```

## 1.5 Інкапсуляція

**Інкапсуляція** — об'єднання даних та методів, що працюють з цими даними, у межах класу, з можливістю приховування внутрішньої реалізації.

### Рівні доступу в Python:



```python
class Example:
    def __init__(self):
        self.public = "Публічний"        # доступний звідусіль
        self._protected = "Захищений"     # конвенція: для внутрішнього використання
        self.__private = "Приватний"      # name mangling

obj = Example()
print(obj.public)       # OK
print(obj._protected)   # OK, але не рекомендується
print(obj.__private)  # AttributeError
```

```python
print(obj._Example__private)
```


### Властивості (Properties):



```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """Getter для температури."""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """Setter з валідацією."""
        if value < -273.15:
            raise ValueError("Температура не може бути нижче абсолютного нуля")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """Обчислювана властивість."""
        return self._celsius * 9/5 + 32

temp = Temperature(25)
print(temp.celsius)      # 25
print(temp.fahrenheit)   # 77.0
temp.celsius = 30        # використання setter
```

## 1.6 Абстрактні класи та інтерфейси

**Абстрактний клас** — клас, який не може бути інстанційований і призначений лише для наслідування.



```python
from abc import ABC, abstractmethod

class AbstractShape(ABC):
    """Абстрактний базовий клас."""
    
    @abstractmethod
    def area(self):
        """Абстрактний метод — МАЄ бути перевизначений."""
        pass
    
    @abstractmethod
    def perimeter(self):
        """Ще один абстрактний метод."""
        pass
    
    def describe(self):
        """Конкретний метод (не абстрактний)."""
        return f"Фігура з площею {self.area()}"

class Square(AbstractShape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4 * self.side

# shape = AbstractShape()  # TypeError: Can't instantiate abstract class
square = Square(5)  # OK
```

## 1.7 Магічні методи

**Магічні методи (dunder methods)** — спеціальні методи з подвійним підкресленням, які визначають поведінку об'єктів.

### Основні магічні методи:



```python
class Point:
    def __init__(self, x, y):
        """Конструктор."""
        self.x = x
        self.y = y
    
    def __str__(self):
        """Рядкове представлення для користувача."""
        return f"Point({self.x}, {self.y})"
    
    def __repr__(self):
        """Рядкове представлення для розробника."""
        return f"Point(x={self.x}, y={self.y})"
    
    def __eq__(self, other):
        """Порівняння на рівність."""
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y
    
    def __add__(self, other):
        """Додавання точок."""
        return Point(self.x + other.x, self.y + other.y)
    
    def __len__(self):
        """Довжина (відстань від початку координат)."""
        return int((self.x**2 + self.y**2)**0.5)

p1 = Point(3, 4)
p2 = Point(1, 2)

print(p1)           # Point(3, 4)
print(p1 + p2)      # Point(4, 6)
print(p1 == p2)     # False
print(len(p1))      # 5
```

де:
* $r$ — радіус кола;
* $a, b, c$ — сторони (прямокутника або трикутника);
* $p$ — півпериметр.

### Тестові дані:
- Коло: r = 5 → S ≈ 78.54
- Прямокутник: a = 4, b = 6 → S = 24
- Трикутник: a = 3, b = 4, c = 5 → S = 6

* **VIP:** 10% знижки + якщо сума > 1000, ще -50 грн ($Total \cdot 0.90 - 50$)

де:
* $Total$ — загальна вартість товарів у кошику без знижок.

### Тестові дані (1500 грн):
- New: 1500
- Regular: 1425
- VIP: 1500 * 0.9 - 50 = 1350 - 50 = 1300

## Варіант 23. "Площі фігур та статистика"

### Опис задачі:
Система обчислення площ різних геометричних фігур та збору статистики.

### Класи:
- **Базовий:** `Shape` (атрибут: `name`)
- **Похідні:** `Square`, `Circle`, `RightTriangle`

### Поліморфний метод:
`area()` — обчислює площу

### Формули:
**Квадрат:**
$$S_{square} = a^2$$

**Коло:**
$$S_{circle} = \pi r^2$$

**Прямокутний трикутник:**
$$S_{triangle} = \frac{1}{2} a \cdot b$$

де:
* $a$ — сторона квадрата або перший катет;
* $b$ — другий катет;
* $r$ — радіус кола.

### Завдання:
1. Створити список фігур.
2. Вивести детальну таблицю.
3. Знайти мін, макс, середню площу.
4. Знайти суму площ, що потрапляють в інтервал $[S_{min}; S_{avg}]$.

### Тестові дані:
- Square a=2 → 4
- Circle r=1 → 3.14159
- Triangle a=3, b=4 → 6

# ДОДАТОК Б. ПРИКЛАД ВИКОНАННЯ (ВАРІАНТ 23)

## Повний лістинг програми



```python
import math

EPS = 1e-9


class Shape:
    """Базовий клас для геометричних фігур."""
    
    def __init__(self, name):
        self.name = name
    
    def area(self):
        """Обчислює площу фігури."""
        raise NotImplementedError("Метод має бути перевизначений")
    
    def params_str(self):
        """Повертає рядок з параметрами фігури."""
        return "-"
    
    def __str__(self):
        """Рядкове представлення."""
        return f"{self.name}: {self.params_str()}, площа={self.area():.6f}"


class Square(Shape):
    """Клас для квадрата."""
    
    def __init__(self, a):
        super().__init__("Квадрат")
        if a <= 0:
            raise ValueError("Сторона має бути додатною")
        self.a = a
    
    def area(self):
        return self.a * self.a
    
    def params_str(self):
        return f"a={self.a:.6f}"


class Circle(Shape):
    """Клас для кола."""
    
    def __init__(self, r):
        super().__init__("Коло")
        if r <= 0:
            raise ValueError("Радіус має бути додатним")
        self.r = r
    
    def area(self):
        return math.pi * self.r * self.r
    
    def params_str(self):
        return f"r={self.r:.6f}"


class RightTriangle(Shape):
    """Клас для прямокутного трикутника."""
    
    def __init__(self, a, b):
        super().__init__("Прямокутний трикутник")
        if a <= 0 or b <= 0:
            raise ValueError("Катети мають бути додатними")
        self.a = a
        self.b = b
    
    def area(self):
        return (self.a * self.b) / 2.0
    
    def params_str(self):
        return f"a={self.a:.6f}, b={self.b:.6f}"


def read_positive_float(prompt):
    """Зчитує додатне дійсне число з валідацією."""
    while True:
        try:
            x = float(input(prompt))
            if x <= 0:
                print("❌ Помилка: значення має бути додатним")
                continue
            return x
        except ValueError:
            print("❌ Помилка: потрібно ввести число")


def read_int_in_range(prompt, lo, hi):
    """Зчитує ціле число в діапазоні з валідацією."""
    while True:
        try:
            n = int(input(prompt))
            if n < lo or n > hi:
                print(f"❌ Помилка: число має бути в діапазоні [{lo}; {hi}]")
                continue
            return n
        except ValueError:
            print("❌ Помилка: потрібно ввести ціле число")


def create_shape(index):
    """Створює фігуру на основі введення користувача."""
    print(f"\n{'─' * 70}")
    print(f"Фігура #{index}")
    print("Оберіть тип:")
    print("  1 — Квадрат")
    print("  2 — Коло")
    print("  3 — Прямокутний трикутник")
    
    shape_type = read_int_in_range("Тип (1/2/3): ", 1, 3)
    
    try:
        if shape_type == 1:
            a = read_positive_float("Введіть сторону a: ")
            return Square(a)
        elif shape_type == 2:
            r = read_positive_float("Введіть радіус r: ")
            return Circle(r)
        else:
            a = read_positive_float("Введіть катет a: ")
            b = read_positive_float("Введіть катет b: ")
            return RightTriangle(a, b)
    except ValueError as e:
        print(f"❌ Помилка створення фігури: {e}")
        return None


def main():
    """Головна функція програми."""
    print("=" * 70)
    print("ПРАКТИЧНА РОБОТА №4: ООП")
    print("Варіант 23: Площі фігур та підсумкова статистика")
    print("=" * 70)
    
    # Введення кількості фігур
    n = read_int_in_range("\nВведіть кількість фігур N (1-100): ", 1, 100)
    
    # Створення колекції фігур
    shapes = []
    for i in range(1, n + 1):
        shape = create_shape(i)
        if shape:
            shapes.append(shape)
    
    if not shapes:
        print("\n❌ Не створено жодної фігури!")
        return
    
    # Обчислення площ
    print("\n" + "=" * 70)
    print(f"{'№':>5} | {'Тип':<26} | {'Параметри':<28} | {'Площа':>12}")
    print("─" * 70)
    
    areas = []
    for idx, shape in enumerate(shapes, start=1):
        s = shape.area()
        areas.append(s)
        print(f"{idx:5d} | {shape.name:<26} | {shape.params_str():<28} | {s:12.6f}")
    
    # Статистика
    s_min = min(areas)
    s_max = max(areas)
    s_sum = sum(areas)
    s_avg = s_sum / len(areas)
    
    # Сума в інтервалі [S_min; S_avg]
    sum_interval = 0.0
    count_interval = 0
    for s in areas:
        if (s_min - EPS) <= s <= (s_avg + EPS):
            sum_interval += s
            count_interval += 1
    
    # Виведення результатів
    print("=" * 70)
    print("СТАТИСТИКА:")
    print("─" * 70)
    print(f"S_min (мінімальна площа) = {s_min:.6f}")
    print(f"S_max (максимальна площа) = {s_max:.6f}")
    print(f"S_avg (середня площа)    = {s_avg:.6f}")
    print(f"S_sum (сума всіх площ)   = {s_sum:.6f}")
    print("─" * 70)
    print(f"Сума площ у інтервалі [S_min; S_avg] = {sum_interval:.6f}")
    print(f"Кількість фігур у інтервалі          = {count_interval}")
    print("=" * 70)


if __name__ == "__main__":
    main()
```

# ДОДАТОК Д. ГЛОСАРІЙ

**Абстрактний клас** — клас, який не може бути інстанційований і призначений лише для наслідування.

**Атрибут** — змінна, що належить об'єкту або класу.

**Базовий клас (батьківський)** — клас, від якого успадковуються інші класи.

**Інкапсуляція** — об'єднання даних та методів в межах класу з можливістю приховування реалізації.

**Клас** — шаблон для створення об'єктів, що визначає їх структуру та поведінку.

**Конструктор** — метод `__init__`, що викликається при створенні об'єкта.

**Магічні методи (dunder)** — спеціальні методи з подвійним підкресленням (\_\_init\_\_, \_\_str\_\_).

**Метод** — функція, що належить класу та визначає поведінку об'єктів.

**Множинне наслідування** — наслідування від кількох батьківських класів одночасно.

**Наслідування** — механізм створення нового класу на основі існуючого.

**Об'єкт (екземпляр)** — конкретна реалізація класу з конкретними значеннями атрибутів.

**ООП** — Об'єктно-орієнтоване програмування.

**Перевизначення (override)** — заміна методу батьківського класу в похідному класі.

**Поліморфізм** — можливість об'єктів різних класів реагувати на однакові виклики по-різному.

**Похідний клас (дочірній)** — клас, створений шляхом наслідування від базового класу.

**Self** — посилання на поточний об'єкт всередині методів класу.

**Super()** — функція для виклику методів батьківського класу.
