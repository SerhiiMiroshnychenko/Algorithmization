## 1 ТЕОРЕТИЧНІ ВІДОМОСТІ

### 1.1 Функції в Python

**Функція** – це незалежна іменована частина програми, яка може багаторазово викликатися з інших частин програми, маніпулювати даними та повертати результати. Кожна функція має власне ім'я, за яким здійснюють її виклик.

Розрізнюють два основні різновиди функцій:

1. **Вбудовані функції** – які є складовою частиною мови програмування або стандартних бібліотек, наприклад: `print()`, `len()`, `abs()`, `math.sin()`, `math.sqrt()` тощо.
2. **Користувацькі функції** – які створюються безпосередньо користувачем для власних потреб.

### 1.2 Оголошення та визначення функцій у Python

У Python функція оголошується та визначається одночасно за допомогою ключового слова `def`:

```python
def function_name(параметри):
    """Документаційний рядок (опціонально)"""
    # тіло функції
    # обчислення
    return результат
```

**Структура функції:**

- **`def`** – ключове слово для оголошення функції
- **`function_name`** – ім'я функції (має бути змістовним)
- **`параметри`** – перелік формальних параметрів через кому (може бути порожнім)
- **`docstring`** – рядок документації, що описує призначення функції (опціонально, але рекомендовано)
- **`return`** – оператор повернення результату (якщо функція не повертає значення, `return` можна не вказувати)

**Приклад простої функції:**

````python
def calculate_square(x):
    """Calculate the square of a number"""
    return x ** 2
````

**Виклик функції:**

````python
result = calculate_square(5)
print(result)  # Виведе: 25
````

### 1.3 Параметри та аргументи функцій

- **Формальні параметри** – змінні, які вказуються при оголошенні функції.
- **Фактичні аргументи** – конкретні значення, які передаються функції при її виклику.

**Приклад:**

````python
def add_numbers(a, b):  # a, b - формальні параметри
    return a + b

sum_result = add_numbers(3, 7)  # 3, 7 - фактичні аргументи
````

### 1.4 Область видимості змінних

У Python змінні мають область видимості:

- **Локальні змінні** – змінні, оголошені всередині функції. Вони доступні тільки всередині цієї функції.
- **Глобальні змінні** – змінні, оголошені поза функціями. Вони доступні у всій програмі.

**Приклад:**

````python
global_var = 10  # глобальна змінна

def example_function():
    local_var = 5  # локальна змінна
    print(f"Global: {global_var}, Local: {local_var}")

example_function()
print(local_var)  # Помилка! local_var не доступна поза функцією
````

### 1.5 Функції з кількома операціями

Функція може виконувати кілька операцій і містити складну логіку:

````python
def process_value(x):
    if x < 0:
        return abs(x)
    elif x == 0:
        return 1
    else:
        return x ** 2
````

### 1.6 Табуляція функцій

**Табуляція функції** – це обчислення значень функції в заданому діапазоні з певним кроком і виведення результатів у вигляді таблиці.

**Загальний алгоритм:**

1. Ввести початкове значення `x_start`, кінцеве значення `x_end` та крок `h`.
2. Створити цикл від `x_start` до `x_end` з кроком `h`.
3. Для кожного значення `x` обчислити значення функції `y = f(x)`.
4. Вивести пару `(x, y)`.

**Приклад простої табуляції:**

````python
import math

def tabulate_function(x_start, x_end, h):
    x = x_start
    print(f"{'x':>10} | {'y':>10}")
    print("-" * 25)
    while x <= x_end:
        y = math.sin(x)
        print(f"{x:10.2f} | {y:10.4f}")
        x += h

tabulate_function(0, math.pi, 0.5)
````

### Приклад 2.1

**Задача:** Описати функцію обчислення модуля дійсного числа та використати її для знаходження максимального за модулем числа серед послідовності.

**Розв'язок:**

````python
def custom_abs(value):
    """Calculate absolute value of a number"""
    if value >= 0:
        return value
    else:
        return -value

# Main program
n = int(input("Введіть кількість чисел: "))
max_abs = float(input("Введіть перше число: "))

for i in range(2, n + 1):
    current = float(input(f"Введіть число {i}: "))
    if custom_abs(current) > custom_abs(max_abs):
        max_abs = current

print(f"Максимальне за модулем число: {max_abs}")
````

### Приклад 2.2

**Задача:** Визначити, чи можна введене з клавіатури число представити у вигляді суми двох простих чисел.

**Розв'язок:**

````python
def is_prime(num):
    """Check if number is prime"""
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Main program
n = int(input("Введіть число: "))
found = False

for i in range(2, n // 2 + 1):
    if is_prime(i) and is_prime(n - i):
        print(f"{n} = {i} + {n - i}")
        found = True
        break

if not found:
    print(f"Число {n} не можна представити як суму двох простих чисел")
````

### Приклад 2.3

**Задача:** Розробити програму для табуляції кусково-заданої функції в інтервалі від `x_start` до `x_end` з кроком `h`, використовуючи користувацькі функції.

Кусково-задана функція:

$$
y(x) = \begin{cases}
x^2 + 1, & \text{якщо } x < -2 \\
\sin(x), & \text{якщо } -2 \leq x < 1 \\
\sqrt{x + 2}, & \text{якщо } x \geq 1
\end{cases}
$$

**Математичний аналіз:**

- **ОДЗ функції:**
  - Для $x < -2$: функція завжди визначена.
  - Для $-2 \leq x < 1$: функція завжди визначена.
  - Для $x \geq 1$: необхідно, щоб $x + 2 \geq 0$, що виконується для всіх $x \geq 1$.

**Блок-схема алгоритму:**

#### Варіант 1: Загальна блок-схема (Mermaid)

```mermaid
flowchart TD
    Start([Початок]) --> Input1[Ввести x_start]
    Input1 --> Input2[Ввести x_end]
    Input2 --> Input3[Ввести h]
    Input3 --> Header[Вивести заголовок таблиці]
    Header --> InitX[x = x_start]
    InitX --> Check{x ≤ x_end + EPS?}
    Check -->|Ні| End([Кінець])
    Check -->|Так| Cond1{x < -2 - EPS?}
    Cond1 -->|Так| Calc1[y = x² + 1]
    Cond1 -->|Ні| Cond2{-2 - EPS ≤ x < 1 - EPS?}
    Cond2 -->|Так| Calc2[y = sin x]
    Cond2 -->|Ні| Calc3[y = √ x + 2]
    Calc1 --> Output[Вивести x, y]
    Calc2 --> Output
    Calc3 --> Output
    Output --> Increment[x = x + h]
    Increment --> Check
```

#### Варіант 2: Загальна блок-схема (PlantUML)

````python
import plantuml
from IPython.display import Image, display

# Створення об'єкта PlantUML
plantuml_code = """
@startuml
start
:Ввести x_start;
:Ввести x_end;
:Ввести h;
:Вивести заголовок таблиці;
:x = x_start;
while (x ≤ x_end + EPS?) is (Так)
  if (x < -2 - EPS?) then (Так)
    :y = x² + 1;
  elseif (-2 - EPS ≤ x < 1 - EPS?) then (Так)
    :y = sin(x);
  else (x ≥ 1 - EPS)
    :y = √(x + 2);
  endif
  :Вивести x, y;
  :x = x + h;
endwhile (Ні)
stop
@enduml
"""

# Рендерінг діаграми
server = plantuml.PlantUML(url='http://www.plantuml.com/plantuml/img/')
png_data = server.get_url(plantuml_code)
display(Image(url=png_data))
````

**Лістінг програми:**

````python
import math

# Константа точності
EPS = 1e-9


def input_data():
    """Input initial data: x_start, x_end, step"""
    x_start = float(input("Введіть x_start: "))
    x_end = float(input("Введіть x_end: "))
    step = float(input("Введіть крок h: "))
    return x_start, x_end, step


def func_y1(x):
    """Calculate y1 = x^2 + 1 for x < -2"""
    y = x ** 2 + 1
    return y


def func_y2(x):
    """Calculate y2 = sin(x) for -2 <= x < 1"""
    y = math.sin(x)
    return y


def func_y3(x):
    """Calculate y3 = sqrt(x + 2) for x >= 1"""
    y = math.sqrt(x + 2)
    return y


def main():
    """Main program"""
    # Введення вихідних даних
    x_start, x_end, step = input_data()
    
    # Виведення заголовка таблиці
    print(f"\n{'x':>10} | {'y':>12}")
    print("-" * 25)
    
    # Табуляція функції
    x = x_start
    while x <= x_end + EPS:
        if x < -2 - EPS:
            # Інтервал 1: x < -2
            y = func_y1(x)
            print(f"{x:10.4f} | {y:12.6f}")
        elif -2 - EPS <= x < 1 - EPS:
            # Інтервал 2: -2 <= x < 1
            y = func_y2(x)
            print(f"{x:10.4f} | {y:12.6f}")
        elif x >= 1 - EPS:
            # Інтервал 3: x >= 1
            y = func_y3(x)
            print(f"{x:10.4f} | {y:12.6f}")
        
        x += step


# Запуск програми
if __name__ == "__main__":
    main()
````

### Постановка задачі

Розробити програму для табуляції кусково-заданої функції $y(x)$ в інтервалі від `x_start` до `x_end` з кроком `h`.

**Варіант 23:**

$$
y(x) = \begin{cases}
\cos(x), & \text{якщо } x < -\frac{\pi}{2} \\
x^2 - 2x + 1, & \text{якщо } -\frac{\pi}{2} \leq x < 2 \\
\ln(x + 3), & \text{якщо } x \geq 2
\end{cases}
$$

### Математична постановка задачі

**ОДЗ функції:**

1. Для $x < -\frac{\pi}{2}$: функція $\cos(x)$ визначена для всіх $x$.
2. Для $-\frac{\pi}{2} \leq x < 2$: функція $x^2 - 2x + 1$ визначена для всіх $x$.
3. Для $x \geq 2$: функція $\ln(x + 3)$ визначена при $x + 3 > 0$, тобто $x > -3$. Оскільки $x \geq 2$, умова виконується автоматично.

**Висновок:** Функція визначена для всіх $x$ з вказаних інтервалів.

### Блок-схема алгоритму

#### Варіант 1: Загальна блок-схема (Mermaid)

### Лістінг програми

````python
import math

# Константа точності
EPS = 1e-9


def input_data():
    """Input initial data: x_start, x_end, step"""
    print("=" * 50)
    print("ТАБУЛЯЦІЯ КУСКОВО-ЗАДАНОЇ ФУНКЦІЇ")
    print("Варіант 23")
    print("=" * 50)
    
    x_start = float(input("Введіть x_start: "))
    x_end = float(input("Введіть x_end: "))
    step = float(input("Введіть крок h: "))
    return x_start, x_end, step


def func_y1(x):
    """Calculate y1 = cos(x) for x < -pi/2"""
    y = math.cos(x)
    return y


def func_y2(x):
    """Calculate y2 = x^2 - 2x + 1 for -pi/2 <= x < 2"""
    y = x ** 2 - 2 * x + 1
    return y


def func_y3(x):
    """Calculate y3 = ln(x + 3) for x >= 2"""
    y = math.log(x + 3)
    return y


def main():
    """Main program"""
    # Введення вихідних даних
    x_start, x_end, step = input_data()
    
    # Виведення заголовка таблиці
    print(f"\n{'=' * 50}")
    print(f"{'x':>12} | {'y':>15} | {'Формула':>18}")
    print(f"{'=' * 50}")
    
    # Табуляція функції
    x = x_start
    while x <= x_end + EPS:
        if x < -math.pi / 2 - EPS:
            # Інтервал 1: x < -π/2
            y = func_y1(x)
            formula = "cos(x)"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        elif -math.pi / 2 - EPS <= x < 2 - EPS:
            # Інтервал 2: -π/2 <= x < 2
            y = func_y2(x)
            formula = "x² - 2x + 1"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        elif x >= 2 - EPS:
            # Інтервал 3: x >= 2
            y = func_y3(x)
            formula = "ln(x + 3)"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        
        x += step
    
    print(f"{'=' * 50}")


# Запуск програми
if __name__ == "__main__":
    main()
````
