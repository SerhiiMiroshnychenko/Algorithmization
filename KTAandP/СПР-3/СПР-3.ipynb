### 1.1 Функції в Python

**Функція** – це незалежна іменована частина програми, яка може багаторазово викликатися з інших частин програми, маніпулювати даними та повертати результати. Кожна функція має власне ім'я, за яким здійснюють її виклик.

Розрізнюють два основні різновиди функцій:

1. **Вбудовані функції** – які є складовою частиною мови програмування або стандартних бібліотек, наприклад: `print()`, `len()`, `abs()`, `math.sin()`, `math.sqrt()` тощо.
2. **Користувацькі функції** – які створюються безпосередньо користувачем для власних потреб.

### 1.2 Оголошення та визначення функцій у Python

У Python функція оголошується та визначається одночасно за допомогою ключового слова `def`:

```python
def function_name(параметри):
    """Документаційний рядок (опціонально)"""
    # тіло функції
    # обчислення
    return результат
```

**Структура функції:**

- **`def`** – ключове слово для оголошення функції
- **`function_name`** – ім'я функції (має бути змістовним)
- **`параметри`** – перелік формальних параметрів через кому (може бути порожнім)
- **`docstring`** – рядок документації, що описує призначення функції (опціонально, але рекомендовано)
- **`return`** – оператор повернення результату (якщо функція не повертає значення, `return` можна не вказувати)

**Приклад простої функції:**

````python
def calculate_square(x):
    """Calculate the square of a number"""
    return x ** 2
````

**Виклик функції:**

````python
result = calculate_square(5)
print(result)  # Виведе: 25
````

### 1.3 Параметри та аргументи функцій

- **Формальні параметри** – змінні, які вказуються при оголошенні функції.
- **Фактичні аргументи** – конкретні значення, які передаються функції при її виклику.

**Приклад:**

````python
def add_numbers(a, b):  # a, b - формальні параметри
    return a + b

sum_result = add_numbers(3, 7)  # 3, 7 - фактичні аргументи
````

### 1.4 Область видимості змінних

У Python змінні мають область видимості:

- **Локальні змінні** – змінні, оголошені всередині функції. Вони доступні тільки всередині цієї функції.
- **Глобальні змінні** – змінні, оголошені поза функціями. Вони доступні у всій програмі.

**Приклад:**

````python
global_var = 10  # глобальна змінна

def example_function():
    local_var = 5  # локальна змінна
    print(f"Global: {global_var}, Local: {local_var}")

example_function()
print(local_var)  # Помилка! local_var не доступна поза функцією
````

### 1.6 Табуляція функцій

**Табуляція функції** – це обчислення значень функції в заданому діапазоні з певним кроком і виведення результатів у вигляді таблиці.

**Загальний алгоритм:**

1. Ввести початкове значення `x_start`, кінцеве значення `x_end` та крок `h`.
2. Створити цикл від `x_start` до `x_end` з кроком `h`.
3. Для кожного значення `x` обчислити значення функції `y = f(x)`.
4. Вивести пару `(x, y)`.

**Приклад простої табуляції:**

````python
import math

def tabulate_function(x_start, x_end, h):
    x = x_start
    print(f"{'x':>10} | {'y':>10}")
    print("-" * 25)
    while x <= x_end:
        y = math.sin(x)
        print(f"{x:10.2f} | {y:10.4f}")
        x += h

tabulate_function(0, math.pi, 0.5)
````

### Приклад 2.1

**Задача:** Описати функцію обчислення модуля дійсного числа та використати її для знаходження максимального за модулем числа серед послідовності.

**Розв'язок:**

````python
def custom_abs(value):
    """Calculate absolute value of a number"""
    if value >= 0:
        return value
    else:
        return -value

# Main program
n = int(input("Введіть кількість чисел: "))
max_abs = float(input("Введіть перше число: "))

for i in range(2, n + 1):
    current = float(input(f"Введіть число {i}: "))
    if custom_abs(current) > custom_abs(max_abs):
        max_abs = current

print(f"Максимальне за модулем число: {max_abs}")
````

### Приклад 2.2

**Задача:** Визначити, чи можна введене з клавіатури число представити у вигляді суми двох простих чисел.

**Розв'язок:**

````python
def is_prime(num):
    """Check if number is prime"""
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Main program
n = int(input("Введіть число: "))
found = False

for i in range(2, n // 2 + 1):
    if is_prime(i) and is_prime(n - i):
        print(f"{n} = {i} + {n - i}")
        found = True
        break

if not found:
    print(f"Число {n} не можна представити як суму двох простих чисел")
````

**Математичний аналіз:**

  - Для $x < -2$: функція завжди визначена.
  - Для $-2 \leq x < 1$: функція завжди визначена.
  - Для $x \geq 1$: необхідно, щоб $x + 2 \geq 0$, що виконується для всіх $x \geq 1$.

#### Рисунок 2 – Блок-схеми функцій обчислення y1, y2 та y3

**Варіант 1: Mermaid**

```mermaid
flowchart TD
    subgraph func1["Функція func_y1(x) - для x < -2"]
        Start1([func_y1 x]) --> Calc1[y = x² + 1]
        Calc1 --> Return1[Повернути y]
        Return1 --> End1([return])
    end
    
    subgraph func2["Функція func_y2(x) - для -2 ≤ x < 1"]
        Start2([func_y2 x]) --> Calc2[y = sin x]

        Return2 --> End2([return])
    end

        Calc3 --> Return3[Повернути y]
        Return3 --> End3([return])
    end
```

**Лістінг програми:**

````python
import math

# Константа точності
EPS = 1e-9


def input_data():
    """Input initial data: x_start, x_end, step"""
    x_start = float(input("Введіть x_start: "))
    x_end = float(input("Введіть x_end: "))
    step = float(input("Введіть крок h: "))
    return x_start, x_end, step


def func_y1(x):
    """Calculate y1 = x^2 + 1 for x < -2"""
    y = x ** 2 + 1
    return y


def func_y2(x):
    """Calculate y2 = sin(x) for -2 <= x < 1"""
    y = math.sin(x)
    return y


def func_y3(x):
    """Calculate y3 = sqrt(x + 2) for x >= 1"""
    y = math.sqrt(x + 2)
    return y


def main():
    """Main program"""
    # Введення вихідних даних
    x_start, x_end, step = input_data()
    
    # Виведення заголовка таблиці
    print(f"\n{'x':>10} | {'y':>12}")
    print("-" * 25)
    
    # Табуляція функції
    x = x_start
    while x <= x_end + EPS:
        if x < -2 - EPS:
            # Інтервал 1: x < -2
            y = func_y1(x)
            print(f"{x:10.4f} | {y:12.6f}")
        elif -2 - EPS <= x < 1 - EPS:
            # Інтервал 2: -2 <= x < 1
            y = func_y2(x)
            print(f"{x:10.4f} | {y:12.6f}")
        elif x >= 1 - EPS:
            # Інтервал 3: x >= 1
            y = func_y3(x)
            print(f"{x:10.4f} | {y:12.6f}")
        
        x += step


# Запуск програми
if __name__ == "__main__":
    main()
````

##### Постановка задачі

Розробити програму для табуляції кусково-заданої функції $y(x)$ в інтервалі від `x_start` до `x_end` з кроком `h`, використовуючи користувацькі функції:

1. Введення значень `x_start`, `x_end` та `h` з клавіатури організувати у вигляді функції.
2. Обчислення елементів кусково-заданої функції організувати у вигляді користувацьких функцій.

##### Кусково-задана функція:

$$
y(x) = \begin{cases}
\frac{\cos(x)}{1 - \tan(x)}, & \text{якщо } x < -3.2 \\
\text{не визначена}, & \text{якщо } -3.2 \leq x \leq -3 \\
\text{sign}(x + \frac{2}{x^2}) \cdot \left|x + \frac{2}{x^2}\right|^{1/5}, & \text{якщо } -3 < x < 0 \\
2^{\cos(x)} + \sin^2(x), & \text{якщо } 0 \leq x \leq 2.5 \\
\text{не визначена}, & \text{якщо } 2.5 < x < 3 \\
\sqrt[6]{x}, & \text{якщо } x \geq 3
\end{cases}
$$

---

#### Математична постановка задачі

##### ОДЗ функції:

###### **Інтервал 1:** $x < -3.2$

Функція: $y_1 = \frac{\cos(x)}{1 - \tan(x)}$

**ОДЗ:**
1. $\cos(x) \neq 0$ (знаменник tan(x))
2. $1 - \tan(x) \neq 0$, тобто $\tan(x) \neq 1$
3. $x \neq \frac{\pi}{2} + \pi k$

###### **Інтервал 2:** $-3.2 \leq x \leq -3$

Функція **не визначена** на цьому інтервалі.

###### **Інтервал 3:** $-3 < x < 0$

Функція: $y_2 = \text{sign}(x + \frac{2}{x^2}) \cdot \left|x + \frac{2}{x^2}\right|^{1/5}$

**ОДЗ:**
1. $x \neq 0$ (знаменник дробу)
2. $x + \frac{2}{x^2} \neq 0$ (для обчислення кореня)

###### **Інтервал 4:** $0 \leq x \leq 2.5$

Функція: $y_3 = 2^{\cos(x)} + \sin^2(x)$

**ОДЗ:** Функція визначена для всіх $x$ з цього інтервалу.

###### **Інтервал 5:** $2.5 < x < 3$

Функція **не визначена** на цьому інтервалі.

###### **Інтервал 6:** $x \geq 3$

Функція: $y_4 = \sqrt[6]{x} = x^{1/6}$

**ОДЗ:** $x \geq 0$ (для парного кореня). Оскільки $x \geq 3$, умова виконується автоматично.

## Лістінг програми (Python)

````python
import math

# Константа точності
EPS = 1e-9


def input_data():
    """Input initial data: x_start, x_end, step"""
    print("=" * 70)
    print("ТАБУЛЯЦІЯ СКЛАДНОЇ КУСКОВО-ЗАДАНОЇ ФУНКЦІЇ")
    print("Приклад 2.4")
    print("=" * 70)
    
    x_start = float(input("Введіть x_start: "))
    x_end = float(input("Введіть x_end: "))
    step = float(input("Введіть крок h: "))
    return x_start, x_end, step


def func_y1(x):
    """
    Calculate y1 = cos(x)/(1-tan(x)) for x < -3.2
    Returns (value, error_message) or (None, error_message)
    """
    try:
        # Перевірка ОДЗ: cos(x) != 0
        if abs(math.cos(x)) < EPS:
            return None, "Ділення на 0 (cos(x)=0)"
        
        # Перевірка ОДЗ: 1 - tan(x) != 0
        tan_x = math.tan(x)
        if abs(1 - tan_x) < EPS:
            return None, "Ділення на 0 (1-tan(x)=0)"
        
        y = math.cos(x) / (1 - tan_x)
        return y, None
    except Exception as e:
        return None, f"Помилка обчислення: {str(e)}"


def func_y2(x):
    """
    Calculate y2 = sign(x + 2/x²) * |x + 2/x²|^(1/5) for -3 < x < 0
    Returns (value, error_message) or (None, error_message)
    """
    try:
        # Перевірка ОДЗ: x != 0
        if abs(x) < EPS:
            return None, "Ділення на 0 (x=0)"
        
        # Обчислення виразу x + 2/x²
        expr = x + 2 / (x ** 2)
        
        # Перевірка на нуль
        if abs(expr) < EPS:
            return 0.0, None
        
        # Обчислення кореня 5-го ступеня зі збереженням знаку
        sign = 1 if expr > 0 else -1
        y = sign * (abs(expr) ** 0.2)
        return y, None
    except Exception as e:
        return None, f"Помилка обчислення: {str(e)}"


def func_y3(x):
    """
    Calculate y3 = 2^cos(x) + sin²(x) for 0 <= x <= 2.5
    Returns (value, error_message) or (None, error_message)
    """
    try:
        y = 2 ** math.cos(x) + math.sin(x) ** 2
        return y, None
    except Exception as e:
        return None, f"Помилка обчислення: {str(e)}"


def func_y4(x):
    """
    Calculate y4 = x^(1/6) for x >= 3
    Returns (value, error_message) or (None, error_message)
    """
    try:
        # Перевірка ОДЗ: x >= 0
        if x < -EPS:
            return None, "x < 0 (корінь парного ступеня)"
        
        y = x ** (1.0 / 6.0)
        return y, None
    except Exception as e:
        return None, f"Помилка обчислення: {str(e)}"


def main():
    """Main program"""
    # Введення вихідних даних
    x_start, x_end, step = input_data()
    
    # Виведення заголовка таблиці
    print(f"\n{'=' * 70}")
    print(f"{'x':>12} | {'y':>18} | {'Інтервал / Помилка':>30}")
    print(f"{'=' * 70}")
    
    # Табуляція функції
    x = x_start
    count = 0
    
    while x <= x_end + EPS:
        # Перевірка на невизначені інтервали
        if -3.2 - EPS <= x <= -3 + EPS:
            print(f"{x:12.6f} | {'---':>18} | {'Функція не визначена':>30}")
        elif 2.5 - EPS < x < 3 - EPS:
            print(f"{x:12.6f} | {'---':>18} | {'Функція не визначена':>30}")
        else:
            # Визначення інтервалу та обчислення
            if x < -3.2 - EPS:
                # Інтервал 1
                y, error = func_y1(x)
                interval = "y1: x < -3.2"
            elif -3 + EPS < x < 0 - EPS:
                # Інтервал 3
                y, error = func_y2(x)
                interval = "y2: -3 < x < 0"
            elif 0 - EPS <= x <= 2.5 + EPS:
                # Інтервал 4
                y, error = func_y3(x)
                interval = "y3: 0 ≤ x ≤ 2.5"
            elif x >= 3 - EPS:
                # Інтервал 6
                y, error = func_y4(x)
                interval = "y4: x ≥ 3"
            else:
                y, error = None, "Невизначений інтервал"
                interval = "---"
            
            # Виведення результату
            if error:
                print(f"{x:12.6f} | {'ПОМИЛКА':>18} | {error:>30}")
            else:
                print(f"{x:12.6f} | {y:18.10f} | {interval:>30}")
                count += 1
        
        x += step
    
    print(f"{'=' * 70}")
    print(f"Успішно обчислено точок: {count}")
    print(f"{'=' * 70}")


# Запуск програми
if __name__ == "__main__":
    main()
````

### Блок-схема алгоритму частинами

Блок-схема алгоритму розділена на дві частини:
1. Рисунок 1 – Блок-схема головної програми та функції введення вихідних даних
2. Рисунок 2 – Блок-схеми функцій обчислення y1, y2 та y3

---

#### Рисунок 1 – Блок-схема головної програми та функції введення вихідних даних

**Варіант 1: Mermaid**

```mermaid
flowchart TD
    subgraph main["Головна програма main()"]
        Start([Початок]) --> CallInput[Викликати input_data]
        CallInput --> GetData[Отримати x_start, x_end, h]
        GetData --> Header[Вивести заголовок таблиці]
        Header --> InitX[x = x_start]
        InitX --> LoopCheck{x ≤ x_end?}
        LoopCheck -->|Так| CheckInterval["Перевірити інтервал<br/>Викликати відповідну функцію"]
        CheckInterval --> Output[Вивести x, y, formula]
        Output --> Increment[x = x + h]
        Increment --> LoopCheck
        LoopCheck -->|Ні| End([Кінець])
    end
    
    subgraph input["Функція input_data()"]
        InputStart([input_data]) --> Print[Вивести заголовок<br/>та варіант]
        Print --> Input1[Ввести x_start]
        Input1 --> Input2[Ввести x_end]
        Input2 --> Input3[Ввести h]
        Input3 --> ReturnValues[Повернути x_start, x_end, h]
        ReturnValues --> InputEnd([return])
    end
```

**Варіант 2: PlantUML**

````python
# Створення об'єкта PlantUML
plantuml_code = """
@startuml
|Головна програма|
start
:Викликати input_data();
:Отримати x_start, x_end, h;
:Вивести заголовок таблиці;
:x = x_start;

repeat
  :Перевірити інтервал;
  :Викликати відповідну функцію;
  :Вивести x, y, formula;
  :x = x + h;
repeat while (x ≤ x_end?) is (Так)
->Ні;

stop

|Функція input_data|
start
:Вивести заголовок та варіант;
:Ввести x_start;
:Ввести x_end;
:Ввести h;
:Повернути x_start, x_end, h;
stop
@enduml
"""

# Рендерінг діаграми
server = plantuml.PlantUML(url='http://www.plantuml.com/plantuml/img/')
png_data = server.get_url(plantuml_code)
display(Image(url=png_data))
````

### Лістінг програми

````python
import math

# Константа точності
EPS = 1e-9


def input_data():
    """Input initial data: x_start, x_end, step"""
    print("=" * 50)
    print("ТАБУЛЯЦІЯ КУСКОВО-ЗАДАНОЇ ФУНКЦІЇ")
    print("Варіант 23")
    print("=" * 50)
    
    x_start = float(input("Введіть x_start: "))
    x_end = float(input("Введіть x_end: "))
    step = float(input("Введіть крок h: "))
    return x_start, x_end, step


def func_y1(x):
    """Calculate y1 = cos(x) for x < -pi/2"""
    y = math.cos(x)
    return y


def func_y2(x):
    """Calculate y2 = x^2 - 2x + 1 for -pi/2 <= x < 2"""
    y = x ** 2 - 2 * x + 1
    return y


def func_y3(x):
    """Calculate y3 = ln(x + 3) for x >= 2"""
    y = math.log(x + 3)
    return y


def main():
    """Main program"""
    # Введення вихідних даних
    x_start, x_end, step = input_data()
    
    # Виведення заголовка таблиці
    print(f"\n{'=' * 50}")
    print(f"{'x':>12} | {'y':>15} | {'Формула':>18}")
    print(f"{'=' * 50}")
    
    # Табуляція функції
    x = x_start
    while x <= x_end + EPS:
        if x < -math.pi / 2 - EPS:
            # Інтервал 1: x < -π/2
            y = func_y1(x)
            formula = "cos(x)"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        elif -math.pi / 2 - EPS <= x < 2 - EPS:
            # Інтервал 2: -π/2 <= x < 2
            y = func_y2(x)
            formula = "x² - 2x + 1"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        elif x >= 2 - EPS:
            # Інтервал 3: x >= 2
            y = func_y3(x)
            formula = "ln(x + 3)"
            print(f"{x:12.6f} | {y:15.8f} | {formula:>18}")
        
        x += step
    
    print(f"{'=' * 50}")


# Запуск програми
if __name__ == "__main__":
    main()
````
